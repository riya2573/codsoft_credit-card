```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
cd = pd.read_csv('creditcard.csv')
cd.head()
```

```python
cd.tail()
```

```python
cd.info()
```

```python
cd.isnull().sum()
```

```python
cd['Class'].value_counts()
```

```python
legit = cd[cd.Class == 0]
fraud = cd[cd.Class == 1]

print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
cd.groupby('Class').mean()
```

```python
legit_sample = legit.sample(n=492)
```

```python
new_cd = pd.concat([legit_sample, fraud], axis=0)
new_cd.head()
```

```python
new_cd.tail()
```

```python
new_cd['Class'].value_counts()
```

```python
new_cd.groupby('Class').mean()
```

```python
X = new_cd.drop(columns='Class', axis=1)
Y = new_cd['Class']

print(X)
print(Y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)

print(X.shape, x_train.shape, x_test.shape)
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
x_train_predication = lr.predict(x_train)
training_data_accuracy = accuracy_score(x_train_predication, y_train)

print('Accuracy on training data :', training_data_accuracy)
```

```python
x_test_predication = lr.predict(x_test)
training_data_accuracy = accuracy_score(x_test_predication, y_test)

print('Accuracy on testing data :', training_data_accuracy)
```
